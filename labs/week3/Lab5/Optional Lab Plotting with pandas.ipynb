```python
# Import Packages
import os

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import pandas as pd
#import earthpy as et

# Add seaborn general plot specifications
sns.set(font_scale=1.5, style="whitegrid")
```

```python
data=pd.read_csv('avg-precip-months-seasons.csv')
data
```

```python
ax1=data.plot(lw=2, color='g', linestyle = ':', marker='.', markersize=20, legend=False, figsize=(15,15))
plt.title("Plot Challenge 1 \nAverage Precipitation")
plt.xlabel("Month")
plt.ylabel("Mean Precipitation (inches)")
default_x_ticks = range(len(data['months']))
plt.xticks(default_x_ticks, data['months'])
plt.show()
```

# Challenge 2: Bar Plot of Precipitation Data

```python
ax2=data.plot.bar(color='g', edgecolor='black', width=0.8, rot=0, legend=False, figsize=(15,15))
plt.title("Plot Challenge 2 \nAverage Precipitation")
plt.xlabel("Month")
plt.ylabel("Mean Precipitation (inches)")
default_x_ticks = range(len(data['months']))
plt.xticks(default_x_ticks, data['months'])
plt.show()
```

# Challenge 3: Figure with Two Subplots of Precipitation Data

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(15, 20))
plt.suptitle("Challenge 3. Overall Figure Title Using the Suptitle Method")

ax1.plot(data['months'], data['precip'], lw=2, color='g', linestyle = ':', marker='.', markersize=20)
ax1.set(title="Line Plot Challenge 1 \nAverage Precipitation", xlabel ="Month", ylabel="Mean Precipitation (inches)")
        
ax2.bar(data['months'], data['precip'], color='g', edgecolor='black', width=0.8)
ax2.set(title="Bar Plot Challenge 2 \nAverage Precipitation", xlabel ="Month", ylabel="Mean Precipitation (inches)")
        
plt.tight_layout()
plt.show()
```

# Bonus Challenge 4: Plot Grouped Data

```python
colors = {"Winter": "lightgrey",
          "Spring": "green",
          "Summer": "darkgreen",
          "Fall": "brown"}

for label, df in data.groupby("seasons"):
    ax1.plot(data['months'], data['precip'],
            "o",
            label=label,
            color=colors[label])
```
