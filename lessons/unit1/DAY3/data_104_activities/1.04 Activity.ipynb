```python
df['state']=df['state'].str.title()
```

```python
def clean_state(x):
    if x in['Tennessee']:
        return 'TE'
    elif x in['Arizona']:
        return 'AR'
    elif x in['California']:
        return 'CA'
    elif x in['Cali']:
        return 'CA'    
    else:
        return x
```

```python
df['state']=list(map(clean_state,df['state']))
```
