## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# SUB PROBLEM 1: Find the total amount spent for each customer
amount_customer = orders.groupby('CustomerID').sum()['amount_spent'].reset_index()
amount_customer
```

```python
# SUB PROBLEM 2: Find the amount_spent values that correspond to quantiles 75% and 95%
amount95 = amount_customer['amount_spent'].quantile(0.95) # This percentile it will correspond to VIP customers
amount75 = amount_customer['amount_spent'].quantile(0.75) # This percentile it will correspond to Preferred customers
```

```python
# SUB PROBLEM 3: Create a funtion to label the customers according to their amount_spent values  
def amount_spent (x):
    if x['amount_spent'] >= amount95:
        return 'VIP'
    elif x['amount_spent'] < amount95 and x['amount_spent'] >= amount75:
        return 'Preferred'
    else:
        return 'Normal'
```

```python
amount_customer['kind']=amount_customer.apply(lambda x: amount_spent(x), axis=1)
```

```python
Cust_Kind=amount_customer[['CustomerID','kind']]
Cust_Kind
```

## Q2: How to identify which country has the most VIP Customers?

```python
#  Merge orders and customers_group_df to have the country and customer group available in a single df
orders_customer = pd.merge(orders, Cust_Kind, on='CustomerID', how='left')
orders_customer
```

```python
# For each customer, find from which countries they made orders
columns = ['CustomerID','Country','kind']

country_customer = pd.DataFrame(orders_customer.groupby(columns).groups.keys(), columns=columns)
country_customer
```

```python
# For all customers, find those who are VIP.
country_vips = country_customer[country_customer['kind']=='VIP']
country_vips
```

```python
# Count the number of customers that made orders from each country
country_count =  country_vips[['Country', 'CustomerID']].groupby('Country').count()
country_count
```

```python
country_count.nlargest(1,'CustomerID')
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
country_vip_pref = country_customer[country_customer['kind'].isin(['VIP','Preferred'])]
country_vip_pref
```

```python
vip_pref_count =  country_vip_pref[['Country', 'CustomerID']].groupby('Country').count()
vip_pref_count
```
